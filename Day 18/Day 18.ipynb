```python
#Library import
import pandas as pd
from math import pi
import matplotlib.pyplot as plt
```

```python
def createRadar(player, data):
    Attributes = ["Defending","Dribbling","Pace","Passing","Physical","Shooting"]
    
    data += data [:1]
    
    angles = [n / 6 * 2 * pi for n in range(6)]
    angles += angles [:1]
    
    ax = plt.subplot(111, polar=True)

    plt.xticks(angles[:-1],Attributes)
    ax.plot(angles,data)
    ax.fill(angles, data, 'blue', alpha=0.1)
    ax.set_yticklabels([])

    ax.set_title(player, va='bottom', fontsize=16, pad=20)

    plt.subplots_adjust(top=0.8)  # Adjust top margin
    
createRadar("Alison",[64,95,93,92,96,92])
```
