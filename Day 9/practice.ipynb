```python
#Import libraries
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#for this practice, we are going to use scikit learn abilities to generate random data
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
#This is to generate 200 data points, with 2 distinct features, and 4 clusters
```

```python
plt.scatter(data[0][:,0],data[0][:,1], c=data[1], cmap="rainbow")
```

```python
data[1]
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4) #inistantiate KMeans with the number of cluster we want
```

```python
kmeans.fit(data[0]) #training the model by fitting the data.
```

```python
kmeans.cluster_centers_ #position of the various cluster centroid
```

```python
kmeans.labels_ #predicted label
```

### *Here we are trying to visualize the difference in clustering between the predicted clusters and the actual clusters, we have the luxury to do this because we are using dummy data generated by scikit learn*

```python
fig, (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize=(12,8))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1], c=kmeans.labels_, cmap="rainbow")

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1], c=data[1], cmap="rainbow")
```
