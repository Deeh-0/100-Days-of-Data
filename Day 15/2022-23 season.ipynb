```python
#import another data set to get insights from
df = pd.read_csv('understat_Liverpool2223.csv')
```

```python
df.head()
```

```python
sns.lineplot(data=df, x='ppda', y='deep')
```

```python
df.isnull().sum() #to check for null values, column wise
```

```python
final_third = df['deep'] #Number of plays in opponent final third
final_third = list(final_third)
```

```python
final_third_allowed = df['deep_allowed'] #Number of plays in Liverpool's final third
final_third_allowed = list(final_third_allowed)
```

```python
pressing = df['ppda']
import ast
def convert_to_dict(pressing):
    dis = ast.literal_eval(pressing)
    return dis['att']/dis['def']

pressing = pressing.apply(convert_to_dict)
pressing = list(pressing)

#LFC pressing stat
```

```python
pressing_allowed = df['ppda_allowed']
import ast
def convert_to_dict(pressing_allowed):
    dis = ast.literal_eval(pressing_allowed)
    return dis['att']/dis['def']

pressing_allowed = pressing_allowed.apply(convert_to_dict)
pressing_allowed = list(pressing_allowed)

#pressing stat against LFC
```

```python
points = df['pts'] 
points = list(points)

#points gotten by LFC in the 22/23 season
```

```python
#save the generate variables into a DataFrame, for ease of use.
lfc_2223 = pd.DataFrame({'points':points,'goals':goals,'goals_allowed':goals_allowed,'shots':shots,'final_third':final_third,'pressing':pressing,'pressing_allowed':pressing_allowed,'possession':possession,'pass_accuracy':pass_accuracy,'final_third_allowed':final_third_allowed,'tackles':tackles,'fouls':fouls,'offside':offside,'saves':saves,'corners':corners,'yellow':yellow,'red':red,})
```
